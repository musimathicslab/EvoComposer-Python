```python
import music21 as m21
```

# Esempio score

```python
sBach = m21.corpus.parse('bach/bwv57.8')
sBach.show("text")
sBach.show()

```

```python
sBach.__dict__
```

```python
# recurse() visita ricorsivamente il contenuto degli oggetti di tipo Stream
for elem in sBach.recurse():
    print(elem)
```

## Accesso alle parti

```python
for part in sBach.parts:
    print(part)

sBach.parts[3].show()
```

```python
bass_stream = m21.stream.Stream()

for note in sBach.parts[3]:
    #print(note)
    bass_stream.append(note)

bass_stream.show()
```
